```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def normal_density(mean, variance, x):
    sigma = np.sqrt(variance)
    return (1 / (np.sqrt(2 * np.pi * variance))) * np.exp(-((x - mean) ** 2) / (2 * variance))

# Example: Plot the normal distribution for different values of µ and σ²
x_values = np.linspace(140, 200, 500)
mean_values = [171, 175]
variance_values = [7.1, 15]

plt.figure(figsize=(8, 5))
for mean, variance in zip(mean_values, variance_values):
    y_values = [normal_density(mean, variance, x) for x in x_values]
    plt.plot(x_values, y_values, label=f"µ={mean}, σ²={variance}")

plt.xlabel("x")
plt.ylabel("Density")
plt.title("Normal Distribution")
plt.legend()
plt.show()
```

```python
def integration(mean, variance, a, b):
    result, _ = quad(normal_density, a, b, args=(mean, variance))
    return result

# Compute probability P(162 ≤ X ≤ 190) for male height distribution
mean = 171
variance = 7.1
probability = integration(mean, variance, 162, 190)
print(f"Probability of height between 162cm and 190cm: {probability:.4f}")
```

```python
def uniform_expectation(a, b):
    return (a + b) / 2

# Example
a, b = 0, 10
print(f"Expectation of Uniform({a}, {b}): {uniform_expectation(a, b)}")
```

```python
def exponential_expectation(lmbda):
    return 1 / lmbda

# Example: Time between pandemics (λ = 1/50)
lmbda = 1/50
print(f"Expected time between pandemics: {exponential_expectation(lmbda)} years")
```

```python
def dosage_function(x):
    return 2.38 * x**2

def expected_dosage(mean, variance):
    result, _ = quad(lambda x: dosage_function(x) * normal_density(mean, variance, x), -np.inf, np.inf)
    return result

# Compute expected dosage
expected_dose = expected_dosage(171, 7.1)
print(f"Expected dosage: {expected_dose:.2f}")
```
