```python
#part a
import sympy as sp

x, y = sp.symbols('x y')
f = sp.exp(x) * sp.sin(y) + y**3

df_dx = sp.diff(f, x)  # Partial derivative w.r.t. x
df_dy = sp.diff(f, y)  # Partial derivative w.r.t. y

df_dx, df_dy
```

```python
#part b
g = x**2 * y + x * y**2

grad_g = [sp.diff(g, x), sp.diff(g, y)]
grad_g_at_point = [expr.subs({x: 1, y: -1}) for expr in grad_g]

# Compute magnitude
magnitude = sp.sqrt(sum(expr**2 for expr in grad_g)).subs({x: 1, y: -1})

grad_g_at_point, magnitude
```

```python
#part c
h = sp.ln(x**2 + y**2)

h_xx = sp.diff(h, x, x)
h_yy = sp.diff(h, y, y)
h_xy = sp.diff(h, x, y)

h_xx, h_yy, h_xy
```

```python
#part d
import numpy as np
import matplotlib.pyplot as plt

j_func = sp.lambdify((x, y), x**3 - 3*x*y + y**3, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title(r'Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```

```python
#part e
k_func = sp.lambdify((x, y), sp.sin(x) * sp.cos(y), 'numpy')

Z_k = k_func(X, Y)

plt.contourf(X, Y, Z_k, levels=50, cmap='plasma')
plt.colorbar()
plt.title(r'Contour plot of $k(x, y) = \sin(x) \cos(y)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```
